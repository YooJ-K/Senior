##LABEL MAP

label_map.pbtxt 작성하기


```python
count = 1
id = []

while True:
    print(count, end="")
    tmp_id = input(' : ')
    
    if tmp_id == '0':
        break
    
    id.append([count, tmp_id])
    count += 1

file_path = '/content/drive/MyDrive/foreign_kindergarten/models/research/object_detection/training/labelmap.pbtxt'
label_file = open(file_path, 'w', encoding='utf-8')
for num, name in id:
    label_file.write('item {\n')
    label_file.write('  id: ' + str(num) + '\n')
    label_file.write('  name: \'+ name + "\'\n")
    label_file.write('}\n')
label_file.close()
```
